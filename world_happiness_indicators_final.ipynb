```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
intellectual_property = "./Project_Resources/Intellectual_Property_Charges.csv"
```

```python
intellectual_property_df = pd.read_csv(intellectual_property)
intellectual_property_df.head()
```

```python
# Create a filtered dataframe from specific columns
intellectual_property_cols = ["Country Code", "Country Name", "2016"]
intellectual_property_transformed= intellectual_property_df[intellectual_property_cols].copy()

# Rename the column headers
intellectual_property_transformed = intellectual_property_transformed.rename(columns={"2016": "Intellectual_Property_Charges", "Country Code": "Country_Code", "Country Name":"Country_Name"})

# Clean the data by dropping duplicates and setting the index
intellectual_property_transformed.set_index("Country_Code", inplace=True)
intellectual_dropna = intellectual_property_transformed.dropna(inplace=True)
print(intellectual_property_transformed)
```

```python
engine = create_engine('mysql+mysqlconnector://root:alAlem@localhost/Happiness_Indicators_2016')
```

```python
# Confirm tables
engine.table_names()
```

```python
intellectual_property_transformed.to_sql(name='intellectual_property_transformed', con=engine, if_exists='append', index=True)
```

# VIRA's WORK

```python
AL = "./Vira_ETL/Arable_Land.csv"
AL_df = pd.read_csv(AL)

AL_df.head()
```

```python
# Create a filtered dataframe from specific columns
AL_cols = ["Country Code", "Country Name", "2016"]
AL_transformed= AL_df[AL_cols].copy()

# Rename the column headers
AL_transformed =AL_transformed.rename(columns={"Country Code": "Country_Code", "Country Name": "Country_Name", "2016": "arable_land",
                                                          })

# Clean the data by dropping duplicates and setting the index
AL_transformed.drop_duplicates("Country_Name", inplace=True)
AL_transformed.set_index("Country_Name", inplace=True)
AL_transformed.dropna(inplace=True)
AL_transformed.head()
```

```python
engine = create_engine('mysql+mysqlconnector://root:alAlem@localhost:3306/Happiness_Indicators_2016')
```

```python
# Confirm tables
engine.table_names()
```

```python
AL_transformed.to_sql(name='arable_land', con=engine, if_exists='append', index=True)
```
