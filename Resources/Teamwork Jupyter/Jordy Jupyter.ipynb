```python
#Importing Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "./Government Education.csv"
original_csv_df = pd.read_csv(csv_file)
original_csv_df.head()
```

```python
#Select columns needed
updated_csv_df = original_csv_df[["Country", "Happiness Rank", "Happiness Score", "Trust (Government Corruption)", "Economy (GDP per Capita)"]].copy()
updated_csv_df
```

```python
final_csv_df = updated_csv_df.rename(columns = {"Country":"Country_Name", "Happiness Rank":"Happiness_Rank",\
                                 "Happiness Score":"Happiness_Score","Trust (Government Corruption)":"Trust_Government_Corruption",\
                                 "Economy (GDP per Capita)":"Economy_GDP_Per_Capita"})

final_csv_df.head()
```

```python
engine = create_engine("mysql+mysqlconnector://root:Js@lguer0@localhost/happiness_indicators_2016")
```
